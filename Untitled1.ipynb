```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#use boxcell Number to make colormap
def MakeColorMap(filename):
    Points = pd.read_csv(filename)
    
    print(Points)
    
    Sx = Points.drop(columns= ['y','z'])
    Sy = Points.drop(columns= ['x','z'])
    
    plt.hist(Sx,bins = 61)
    plt.title("Sx")
    plt.xlabel("x")
    plt.ylabel("the number of hit broadbeam")
    plt.savefig("x.png")
    plt.show()
    plt.close()
    plt.hist(Sy,bins = 61)
    plt.title("Sy")
    plt.xlabel("y")
    plt.ylabel("the number of hit broadbeam")
    plt.savefig("y.png")
    plt.show()
    plt.close()
    
    boxcell = np.zeros((61,61))

    cell_size = int(31/61) 

    for i in range(61):
        for j in range(61):
            x = i - cell_size/2
            y = j - cell_size/2
            
            Points_1 = Points[Points['x'] <= x + 1]
            Points_2 = Points_1[Points_1['x'] > x]
            Points_3 = Points_2[Points_2['y'] <= y+1]
            Points_4 = Points_3[Points_3['y'] > y]
            
            print(Points_1)
            boxcell[i,j] = len(Points_4)
    
    return boxcell
```

```python
cell_broad = MakeColorMap('DitPts_broadbeam.csv')
print(cell_broad)
plt.imshow(cell_broad)
plt.colorbar()
```

```python
Point_pen = MakeColorMap("SurPts_pencilbeam.csv")
plt.imshow(Point_pen)
plt.colorbar()
```
