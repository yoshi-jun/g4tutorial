```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def MakeDoseList_z(filename,Z):

    data_origin = pd.read_csv(filename)
    
    data_dose = data_origin[data_origin['dose']>0].drop(columns = ["ix","iy","iz"])
    data_z = data_origin[data_origin['dose']>0].drop(columns = ["ix","iy","dose"])
    
    deps = np.zeros(Z)
    list_z = np.zeros(Z)
    
    for i in range(len(data_z)):
        deps[data_z.iloc[i]] += data_dose.iloc[i]
    for i in range(Z):
        list_z[i] = 0.2 * i

    return list_z,deps
```

```python
z,deps = MakeDoseList_z("vgeo/g4bench-10,000times.csv",150)
```

```python
def MakeDitSurColorMap(filename):
    test2 = pd.read_csv(filename)
    test2 = test2.drop(columns = "z")
    box = np.zeros((100,100))
    cell = 30.5 / 100
    for i in range(100):
        for j in range(100):

            test_1 = test2[test2["x"] >= i * cell -15.25]
            test_2 = test_1[test_1["x"] < (i+1) * cell -15.25]
            test_3 = test_2[test_2["y"] >= j * cell -15.25]
            test_4 = test_3[test_3["y"] < (j+1) * cell -15.25]

            box[i,j] = len(test_4)
    return box
```

```python
bb = MakeDitSurColorMap("/home/yoshijun/Work/tutorial01/TestBench/data/DitPts_broadbeam.csv")
plt.imshow(bb)
plt.xlabel("x boxcell Number")
plt.ylabel("y boxcell Number")
plt.colorbar()
plt.savefig("DitSurPt_bro(100,100).png")
plt.show()
```
